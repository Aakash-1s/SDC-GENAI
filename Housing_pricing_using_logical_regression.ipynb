```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Step 1: Generate a synthetic dataset for housing prices
np.random.seed(42)

# Example features: [Square footage, Number of rooms]
X = np.random.rand(100, 2) * np.array([2000, 5])  # 100 houses, square footage (0-2000), rooms (0-5)
y = 100000 + X[:, 0] * 200 + X[:, 1] * 50000 + np.random.randn(100) * 50000  # Price = base + (sqft * 200) + (rooms * 50000) + noise

# Step 2: Create binary classification target (1 for above median price, 0 for below median price)
median_price = np.median(y)
y_binary = (y > median_price).astype(int)  # 1 if price is above median, else 0

# Step 3: Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Step 4: Initialize the Logistic Regression model
log_reg = LogisticRegression()

# Step 5: Train the model on the training data
log_reg.fit(X_train, y_train)

# Step 6: Make predictions on the test data
y_pred = log_reg.predict(X_test)

# Step 7: Evaluate the model (accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Step 8: Visualize the results (for square footage vs binary prediction)
plt.scatter(X_test[:, 0], y_test, color='blue', label='Actual Prices (Above/Below Median)')
plt.scatter(X_test[:, 0], y_pred, color='red', label='Predicted (Above/Below Median)')
plt.xlabel('Square Footage')
plt.ylabel('Above Median Price (1) or Below (0)')
plt.legend()
plt.title('Logistic Regression: Housing Price Prediction (Above or Below Median)')
plt.show()
```
